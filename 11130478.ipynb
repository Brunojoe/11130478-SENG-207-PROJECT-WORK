```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset= pd.read_csv('red_wine.csv')
```

```python
# Previewing the first 10 rows using the head method
dataset.head(10)
```

```python
# Previewing the last 20 rows using the tail method
dataset.tail(20)
```

```python
#  to check basic information on the dataset
dataset.info()
```

```python
# to obtain statistical information on the dataset
dataset.describe()
```

```python
x=dataset.alcohol
y=dataset.pH
plt.scatter(x,y)
plt.title('A SCATTER PLOT OF ALCOHOL AGAINST pH')
plt.xlabel('Alcohol level')
plt.ylabel('pH values')
plt.show()
```

```python
x=dataset.index
y=dataset.density
plt.bar(x,y)
plt.title('A BAR GRAPH OF THE WINES AGAINST THEIR DENSITIES')
plt.xlabel('Red wine')
plt.ylabel('Density')
plt.show()
```

## Analysis report ##
__Scatter plot: Alcohol level against pH values__
---
The scatter plot visualizes the relationship between alcohol levels (x-axis) and pH values (y-axis) in the dataset.There appears to be a scattered distribution with no clear linear correlation between alcohol levels and pH values. However, there might be a subtle trend suggesting lower pH values with higher alcohol content, indicating a possible inverse relationship.
## __Bar Graph: Wine Densities__
The bar graph represents the densities of different red wines in the dataset. Each bar corresponds to a specific red wine, showcasing the variability in wine densities. Some wines exhibit higher densities compared to others, indicating a diverse range of density values among the wines in the dataset.

In summary, the scatter plot depicting alcohol levels against pH values demonstrates a scattered distribution with a potential but not definitive inverse relationship between alcohol content and pH. On the other hand, the bar graph highlighting wine densities showcases the variability and range of densities among the different red wines in the dataset.

These visualizations provide insights into the relationships and variations in wine attributes, aiding in understanding the dataset's characteristics and potential trends between alcohol content, pH values, and wine densities.
